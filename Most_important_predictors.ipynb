```python
import pandas as pd
import numpy as np
import tqdm
import os
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

os.chdir('C:/Users/NemethR/PycharmProjects/depression/')

df7 = pd.read_pickle('c:/Users/NemethR/PycharmProjects/depression/data/interim/dataframes/df_pickled_final')
df7=df7.dropna()

of = open('models/logreg/most_important_features_char_ngrams.tsv','w',encoding='utf-8')

############################
#       multinomial LOGREG      #
#   Texts+Features,Tfidf: 2-5 char ngrams   #
############################

classifier = LogisticRegression(multi_class='multinomial', solver='newton-cg', max_iter=10000)
array = df7.values

x_txt = df7.final_txt.values.astype('U')
x_features = array[:,8:28]

y = array[:,2]
y = y.astype('int')

column_names = df7.columns.values
column_names
#nevek = np.array(vektor.get_feature_names())

vektor = TfidfVectorizer(min_df=5, ngram_range=(2,5),
                             analyzer='char_wb').fit(x_txt)
x_vect = vektor.transform(x_txt)
X = np.column_stack((x_vect.A, x_features))
nevek = np.array(vektor.get_feature_names())
uj=['noun','verb','adjective','lexdiv', 'sentfre','wordfre','atl_mhossz','stopnum',
    'stopar','senti','emojiarany','misspelled','szam','szo','veszar','irjar','szohossz',
    'drugfre','rovidites','dosis']
nevek_uj=np.concatenate((nevek,uj),axis=0)


classifier.fit(X, y)
classifier.coef_.shape

# most important features for the first (0) label:
sorted_coef_index = classifier.coef_[0].argsort()
# smallest/largest coefs
# smallest ascending
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
# largest descending
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the second label:
sorted_coef_index = classifier.coef_[1].argsort()
# smallest/largest coefs
# smallest ascending
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
# largest descending
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the 3rd label:
sorted_coef_index = classifier.coef_[2].argsort()
# smallest/largest coefs
# smallest ascending
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
# largest descending
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the 4th label:
sorted_coef_index = classifier.coef_[3].argsort()
# smallest/largest coefs
# smallest ascending
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
# largest descending
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))
of.close()
of = open('models/logreg/most_important_features_word_ngrams.tsv','w',encoding='utf-8')

############################
#       multinomial LOGREG      #
#   Texts+Features,Tfidf: 2-5 char ngrams   #
############################

classifier = LogisticRegression(multi_class='multinomial', solver='newton-cg', max_iter=10000)
array = df7.values

x_txt = df7.final_txt.values.astype('U')
x_features = array[:,8:28]

y = array[:,2]
y = y.astype('int')

column_names = df7.columns.values
column_names
#nevek = np.array(vektor.get_feature_names())

vektor = TfidfVectorizer(min_df=5, ngram_range=(1,3)).fit(x_txt)
x_vect = vektor.transform(x_txt)
X = np.column_stack((x_vect.A, x_features))
nevek = np.array(vektor.get_feature_names())
uj=['noun','verb','adjective','lexdiv', 'sentfre','wordfre','atl_mhossz','stopnum',
    'stopar','senti','emojiarany','misspelled','szam','szo','veszar','irjar','szohossz',
    'drugfre','rovidites','dosis']
nevek_uj=np.concatenate((nevek,uj),axis=0)


classifier.fit(X, y)

classifier.coef_.shape

# most important features for the 1st label:
sorted_coef_index = classifier.coef_[0].argsort()
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the 2nd label:
sorted_coef_index = classifier.coef_[1].argsort()
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the 3rd label:
sorted_coef_index = classifier.coef_[2].argsort()
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))

# most important features for the 4th label:
sorted_coef_index = classifier.coef_[3].argsort()
print('Smallest Coefs:\n{}\n'.format(nevek_uj[sorted_coef_index[:20]]))
print('Largest Coefs: \n{}'.format(nevek_uj[sorted_coef_index[:-21:-1]]))
of.close()
```
